```python
class DICTOFCITIES(object):
    def __init__(self)->CLASSINIT:
        self.maindirectory = "/kaggle/input/feb-06-23earthquake-turkey-citiesvectors/vectors"
    def __str__(self)->str:
        return "DICTIONARY OF CITIES FOR DATA EMBEDDING - SUBPROCESS"
    def __call__(self)->None:
        return None
    def __getstate__(self)->CLASSINIT:
        raise TypeError(stateerror)
    def __repr__(self)->str:
        return DICTOFCITIES.__doc__
    def _EMBED(self,ct:str,bt:str,bt2:str,mty:str,ty:str,dc:dict)->PARAMETERS:
        for r,d,f in os.walk(self.maindirectory):
            for n in f:
                if (ct in os.path.join(r,n)) and (bt in os.path.join(r,n) or bt2 in os.path.join(r,n)) and (mty in os.path.join(r,n)) and (ty in os.path.join(r,n)):
                    dc[ct.upper()].append(os.path.join(r,n))
                else:
                    pass
    def _LAUNCH(self,bt:str,bt2:str,mty:str,ty:str,dc:dict)->RESULTS:
        for xkey in dc.keys():
            self._EMBED(xkey,bt,bt2,mty,ty,dc)
```

```python
class BASEDONCOORDINATE(object):
    def __init__(self)->CLASSINIT:
        pass
    def __str__(self)->str:
        return "COORDINATES & FEATURES - SUBPROCESS"
    def __call__(self)->None:
        return None
    def __getstate__(self)->CLASSINIT:
        raise TypeError(stateerror)
    def __repr__(self)->str:
        return BASEDONCOORDINATE.__doc__
    def _RETURNCOORDINATE(self,
                          lparams:list or tuple,
                          listlat:list,
                          listlon:list)->RESULTS:
        lenparams = len(lparams)
        cntparams = len(lparams[0][3])
        if cntparams == 2:
            for xl in range(lenparams):
                    latx = lparams[xl][3][0]
                    lonx = lparams[xl][3][1]
                    listlat.append(latx)
                    listlon.append(lonx)
        else:
            pass
    def _RETURNZIPCOORDINATES(self,
                             lparams:list or tuple,
                             listlat:list,
                             listlon:list)->RESULTS:
        lt,ln = self._RETURNCOORDINATE(lparams,listlat,listlon)
        return list(zip(lt,ln))
    def _RETURNALLATTRIBUTES(self,
                            lparams:list or tuple,
                            nmtarget:str)->RESULTS:
        nmatt = []
        tyatt = []
        dmatt = []
        ltlat = []
        ltlon = []
        lenparams = len(lparams)
        cntparams = len(lparams[0][3])
        if cntparams == 2:
            for xl in range(lenparams):
                nmatt.append(lparams[xl][0])
                tyatt.append(lparams[xl][1])
                dmatt.append(lparams[xl][2])
                latx = lparams[xl][3][0]
                lonx = lparams[xl][3][1]
                ltlat.append(latx)
                ltlon.append(lonx)
            nseries = pd.Series(nmatt)
            tseries = pd.Series(tyatt)
            dseries = pd.Series(dmatt)
            ltseries = pd.Series(ltlat)
            lnseries = pd.Series(ltlon)
            fr = {"BUILDING_NAME":nseries,
                 "BUILDING_TYPE":tseries,
                 "DAMAGED_TYPE":dseries,
                 "LATITUDE":ltseries,
                 "LONGITUDE":lnseries}
            rs = pd.DataFrame(fr)
            rs.to_csv(f"{nmtarget}.csv")
            return rs
        else:
            pass
```

```python
class RETURNMAP(object):
    def __init__(self)->CLASSINIT:
        self.__lc = [38,35]
        self.__zm = 5
        self.__tl = "Stamen Terrain"
        self.__ml = "/MapServer/tile/{z}/{y}/{x}"
    def __str__(self)->str:
        return "RETURN MAP - SUBPROCESS"
    def __call__(self)->None:
        return None
    def __getstate__(self)->CLASSINIT:
        raise TypeError(stateerror)
    def __repr__(self)->str:
        return RETURNMAP.__doc__
    def _GETSAMPLE(self)->RESULTS:
        mf = folium.Map(tiles=self.__tl,
                       location=self.__lc,
                       zoom_start=self.__zm,
                       attr="<p>EQ - RESEARCH</p>")
        return mf
    def _INITLAYER(self,mp:classmethod)->RESULTS:
        for _xn,_xt in tilelayers.items():
            _xt += self.__ml
            folium.TileLayer(tiles=_xt,
                            name=_xn,
                            attr=f"<p>{_xn.upper()} TYPE - EQ</p>").add_to(mp)
        folium.LayerControl().add_to(mp)
        mp.add_child(folium.LatLngPopup())
        mp.add_child(plugins.Geocoder())
        return mp
    def _LOCAL(self,
              mp:classmethod,
              dp:classmethod,
              cause:str,
              coor:list)->RESULTS:
        for xcount,xtarget in enumerate(coor):
            ifr = f"""
            <html>
            <h6>TARGET: {cause}</h6>
            <div>
            <p>
            BUILDING NAME:
            </p>
            <p>
            <b>{dp.BUILDING_NAME.values[xcount]}</b>
            </p>
            <p>
            DAMAGED:
            </p>
            <p>
            <b>{dp.DAMAGED_TYPE.values[xcount]}</b>
            </p>
            <p>
            COORDINATES:
            </p>
            <p>
            <b>{str(round(xtarget[0],3))}, {str(round(xtarget[1],3))}</b>
            </p>
            </div>
            </html>
            """
            ipop = folium.Popup(folium.IFrame(ifr),
                               min_width=200,
                               max_width=200)
            icn = folium.Icon(color="red",icon="glyphicon-flag")
            folium.Marker(location=[xtarget[0],xtarget[1]],
                         popup=ipop,
                         tooltip="Check",
                         icon=icn).add_to(mp)
    def _ALLRUN(self,
               dp:classmethod,
               cause:str,
               coor:list)->RESULTS:
        mapbase = self._GETSAMPLE()
        self._LOCAL(mapbase,dp,cause,coor)
        mnew = self._INITLAYER(mapbase)
        return mnew
```

```python
class ANALYSISCITIES(object):
    def __init__(self)->CLASSINIT:
        pass
    def __str__(self)->str:
        return "ANALYSIS FOR CITIES - SUBPROCESS"
    def __call__(self)->None:
        return None
    def __getstate__(self)->CLASSINIT:
        raise TypeError(stateerror)
    def __repr__(self)->str:
        return ANALYSISCITIES.__doc__
    def _HISTOGRAM(self,
                   dt:classmethod,
                   xval:str,
                   stitle:str,
                   clr:str,
                   tmp:str="plotly_dark")->RESULTS:
        fg = px.histogram(dt,
                         xval,
                         color=clr,
                         title=f"<b>{stitle}</b>",
                         template=tmp)
        fg.update_traces(textposition='inside',
                          textfont_size=8)
        fg.update_layout(bargap=0.2)
        fg.show()
    def _JUSTTOTAL(self,dt:classmethod,compm:str):
        return len(dt[compm].values)
    def _PERCENTAGE(self,
                   dt:classmethod,
                   compm:str,
                   comp1:str)->RESULTS:
        totallen = len(dt[compm].values)
        targetlen = len(dt[dt[compm] == comp1])
        perc = f"[ {100 * (targetlen/totallen)} ] % WAS [ {comp1.upper()} ]"
        cmtx = f"TOTAL: [ {targetlen} ] WAS [ {comp1.upper()} ]"
        return perc,cmtx
    def _ALLTOTAL(self,
                 dt:classmethod,
                 compm:str,
                 comp1:str,
                 lt:int)->RESULTS:
        totallen = len(dt[compm].values)
        targetlen = len(dt[dt[compm] == comp1])
        lt += targetlen
        return lt
```

```python
for xnme in namecitylist:
    bls = pd.read_csv(f"/content/{xnme}.csv")
    ANALYSISCITIES()._HISTOGRAM(bls,"DAMAGED_TYPE",xnme.upper(),"BUILDING_TYPE")
```

```python
for xnme in namecitylist:
    bls = pd.read_csv(f"/content/{xnme}.csv")
    print(f"\nFOR {xnme.upper()}\n")
    tpl = ANALYSISCITIES()._JUSTTOTAL(bls,"DAMAGED_TYPE")
    print(f"TOTAL BUILDINGS OBSERVATIONS --> {tpl}\n")
    for xtarval in sllist:
        res,cmt = ANALYSISCITIES()._PERCENTAGE(bls,"DAMAGED_TYPE",xtarval)
        print(f"RESULTS:\n\n{cmt}\n{res}\n{'---'*7}\n")
    print("###########"*7)
```

```python
ltdes = 0
ltpde = 0
ltdam = 0
for xnme in namecitylist:
    bls = pd.read_csv(f"/content/{xnme}.csv")
    for xtarval in sllist:
        if xtarval == "Destroyed":
            ltdes = ANALYSISCITIES()._ALLTOTAL(bls,"DAMAGED_TYPE",xtarval,ltdes)
        elif xtarval == "Possibly damaged":
            ltpde = ANALYSISCITIES()._ALLTOTAL(bls,"DAMAGED_TYPE",xtarval,ltpde)
        elif xtarval == "Damaged":
            ltdam = ANALYSISCITIES()._ALLTOTAL(bls,"DAMAGED_TYPE",xtarval,ltdam)
        else:
            pass
print(f"DESTROYED TOTAL: {ltdes}")
print(f"POSSIBLY DAMAGED TOTAL: {ltpde}")
print(f"DAMAGED TOTAL: {ltdam}")
print(f"BADLY AFFECTED PLACES TOTAL: {ltdes+ltpde+ltdam}")
```
